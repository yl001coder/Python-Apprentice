## Nested for-loops

Write nested for loops (a for loop inside another for loop). Here is an example of nested loops. 

````python
# Examples of Nested Loops

for i in range(8):
    for j in range(8):
        print(f"({i},{j})", end= ' ')
    print()
````

````python
# Display the following shown above
for i in range(3):
    for j in range(3):
        print(f"{i} {j}")
    print()
````

