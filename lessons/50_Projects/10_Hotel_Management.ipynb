# Hotel Management

For this project, your goal is to write a program straight from scratch. 

### Requirements
- Should run in 1 While Loop

- Nearly all functionality should be in functions, and called from a
  singular `main()` function.

- A majority of the functions should have a return
  * For instance, if I had a function called add, it should return the sum rather than just doing the operation or printing it.
  * Remember, with Python you can set a variable to a return value of a
  function, so:  my_age = add(10, 5)

- Add arguments to your functions as needed

- Must contain and iterate over these data structures:
  * List
  * Dictionary
  * Tuple

### Functions
- multi-day check-in

- check people in and out of the hotel

- interactive (either through Tkinter or the console using print())

- charge guests for room(s)

- multi-room check-in

- one extra feature of your choice
